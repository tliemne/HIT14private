```python
def cost_function(price,real_price):
  res = real_price -price
  return (0.5*(sum((res)**2)))/len(price)
```

```python
def predict(x):
  one = np.ones((x.shape[0], 1))
  xbar = np.concatenate((one,x), axis = 1)
  A = xbar.T @ xbar
  b = np.array(xbar.T @ train['price'])
  if np.linalg.det(A) != 0:
    return np.linalg.inv(A) @ b
  else:
    return np.linalg.pinv(A) @ x.T @ train['price']
```
